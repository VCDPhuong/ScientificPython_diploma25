```python
import numpy as np
def checkempty(func):
  def wrap(*arg,**kwarg):
    arg = np.asarray(arg)
    vs = np.asarray([v for k, v in kwarg.items()])
    if sum([i.size for i in arg])==0 and sum([i.size for i in vs])==0:
      print('nope')
    else:
      print("yup")
      return func(*arg,**kwarg)
  return wrap
@checkempty
def plus(*B,**C):
  return print(sum(B) + sum(np.asarray([v for k,v in C.items()])))

M1= []
M2= []
M3= [2,2]
M4= [4,4]

plus(M1,M2)
plus(M3,M4)
plus(M5=[],M6=[])
```

```python
import time
def sl(func):
 def wrap(*arg,**kwarg):
  start = time.time()
  func(*arg,*kwarg)
  finish = time.time()
  print('time cost ', finish - start, ' s')
 return wrap

@sl
def add(a,b):
  time.sleep(3)
  print(a+b)

add(3,4)
```

```python
class cpl():
  def __init__(self,c,d):
    self.a = c
    self.b = d
  def __add__(self,other):
    x = self.a + other.a
    y = self.b + other.b
    return cpl(x,y)
  def __minus__(self,other):
    x = self.a - other.a
    y = self.b - other.b
    return cpl(x,y)
  def __mul__(self,other):
    x = (self.a)*other.a - (other.b)*self.b
    y = self.a*other.b + self.b*other.a
    return cpl(x,y)
  def __norm__(self,other):
    n = (self.a + other.a)**2 + (self.b + other.b)**2
    return cpl(n,0)
  def __adj__(self,other):
    return cpl(self.a + other.a,-self.b - other.b)
  def __dev__(self,other):
    x = self.__mul__(self.__adj__(other))
    n = self.__norm__(other)
    return cpl(x.a/n.a,x.b/n.a)
  def __str__(self):
    return "{0},{1}".format(self.a, self.b) #"converts" class to a string


C= cpl(1,3)
D= cpl(1,5)
print(C.__dev__(D))
```
